```python
import torch
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline



```

```python

# Load BioBERT model fine-tuned for NER
tokenizer = AutoTokenizer.from_pretrained("dmis-lab/biobert-v1.1")
model = AutoModelForTokenClassification.from_pretrained("dmis-lab/biobert-v1.1")

# Load NER pipeline
nlp_ner = pipeline("ner", model=model, tokenizer=tokenizer, aggregation_strategy="simple")
```

```python

# Medical entity categories (you may need a fine-tuned model to get specific labels)
CATEGORY_MAPPING = {
    "DISEASE": "Disease",
    "SYMPTOM": "Symptom",
    "TREATMENT": "Cure"
}
```

```python

def extract_medical_terms(text):
    entities = nlp_ner(text)
    categorized_terms = {"Disease": [], "Symptom": [], "Cure": []}

    for entity in entities:
        label = entity['entity_group'].upper()
        word = entity['word']

        if label in CATEGORY_MAPPING:
            categorized_terms[CATEGORY_MAPPING[label]].append(word)

    return categorized_terms
```

```python

# Example user input
user_input = "The patient is suffering from diabetes and experiencing frequent urination. The prescribed medication is Metformin."

# Extract and categorize terms
medical_terms = extract_medical_terms(user_input)
print(medical_terms)
```

```python
def extract_medical_terms(text):
    entities = nlp_ner(text)
    categorized_terms = {"Disease": [], "Symptom": [], "Cure": []}

    # Group multi-word entities and reconstruct subword tokens
    grouped_entities = []
    current_entity = None

    for entity in entities:
        label = entity['entity_group']
        word = entity['word']
        start = entity['start']
        end = entity['end']

        # If the current entity is None or the label changes, start a new entity
        if current_entity is None or current_entity['entity_group'] != label:
            if current_entity is not None:
                grouped_entities.append(current_entity)
            current_entity = {
                'entity_group': label,
                'word': word,
                'start': start,
                'end': end
            }
        else:
            # Append to the current entity and reconstruct subword tokens
            current_entity['word'] += word.replace("##", "")  # Remove "##" for subword tokens
            current_entity['end'] = end

    # Add the last entity
    if current_entity is not None:
        grouped_entities.append(current_entity)

    # Categorize the grouped entities
    for entity in grouped_entities:
        label = entity['entity_group']
        word = entity['word']

        # Map using the label directly (adjust according to model's labels)
        if label in CATEGORY_MAPPING:
            category = CATEGORY_MAPPING[label]
            categorized_terms[category].append(word)
        else:
            print(f"Ignored entity: {word} (Label: {label})")  # Debug ignored entities

    return categorized_terms
```

```python
# Example user input
user_input = "The patient is suffering from diabetes and experiencing frequent urination. The prescribed medication is Metformin."

# Extract and categorize terms
medical_terms = extract_medical_terms(user_input)
print("Categorized terms:", medical_terms)
```
